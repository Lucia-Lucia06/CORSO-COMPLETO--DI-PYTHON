```python
import pandas as pd
import matplotlib.pyplot as plt

dipendenti = {
    "nome": ["Mario", "Rudy", "Marco", "Luca", "Giovanni", "Francesco", "Giulia", "Elena", "Federica", "Alessandro"],
    "cognome": ["Ricci", "Botosso", "Secci", "Bianchi", "Neri", "Rossi", "Verdi", "Gialli", "Blu", "Rosa"],
    "email": ["mario.ricci@example.info", None, "marco.secci@example.com", None, "giovanni.neri@example.com", "francesco.rossi@example.ue", "giulia.verdi@example.com", "elena.gialli@example.it", "federica.blu@example.com", "alessandro.rosa@example.info"],
    "eta": [30, 54, 54, 28, 35, 42, 28, 32, 29, 31],
    "citta": ["Milano", "Biella", "San Donà di Piave", "Roma", "Napoli", "Torino", "Firenze", "Bologna", "Genova", "Padova"],
    "stipendio": [2000, 3000, 1500, 2500, 3500, 4000, 2500, 3000, 3500, 4000],
    "ruolo": ["Sviluppatore", "Project Manager", "Analista", "Sviluppatore", "Team Leader", "Project Manager", "Sviluppatore", "Analista", "Team Leader", "Project Manager"],
    "data_assunzione": ["2015-01-15", "2010-06-30", "2018-09-01", "2016-03-20", "2012-11-05", "2008-04-10", "2017-07-25", "2014-02-18", "2013-12-12", "2009-08-01"]
}

df = pd.DataFrame(dipendenti)
df
```

## Grafico a barra: Stipendio per dipendenti 



```python
plt.figure(figsize=(12, 6)) # Dimensione della finestra
plt.bar(df['nome'], df['stipendio']) # SELECT nome, stipendio from Dipendenti 
plt.show() # Visualizzazione della figura
```

```python
plt.figure(figsize=(12, 6)) # Dimensione della finestra
plt.title('Stipendio dei dipendenti') # Etichetta del titolo
plt.barh(df['nome'], df['stipendio']) # SELECT nome, stipendio from Dipendenti 
plt.show() # Visualizzazione della figura
```

```python
plt.figure(figsize=(12, 6)) # Dimensione della finestra
plt.title('Stipendio dei dipendenti') # Etichetta del titpolo
plt.bar(df['nome'], df['stipendio'],color='purple') # SELECT nome, stipendio from Dipendenti
plt.xlabel('Nome') # Titolo orizzontale
plt.ylabel('Stipendio (€)') # Titolo verticale
plt.xticks(rotation=45) # Rotazione nomi orizzontale
plt.yticks(rotation=45) # Rotazione nomi verticale
plt.show() # Visualizzazione della figura
```

## Grafico a torta: Ruoli aziendali

```python
conteggio_ruoli = df['ruolo'].value_counts()

plt.figure(figsize=(12, 6))
plt.pie(conteggio_ruoli)
plt.show()
```

```python
conteggio_ruoli = df['ruolo'].value_counts()

plt.figure(figsize=(12, 6))
plt.pie(conteggio_ruoli, labels=conteggio_ruoli.index, autopct='%1.2f%%', startangle=90)
plt.show()
```

```python
conteggio_ruoli = df['ruolo'].value_counts()

plt.figure(figsize=(12, 6))
plt.title('Ruoli dei Dipendenti')
plt.pie(conteggio_ruoli, labels=conteggio_ruoli.index, autopct='%1.2f%%', startangle=90)
plt.show()
```
